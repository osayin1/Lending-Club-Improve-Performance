## Purpose: Developing a Predictive Model that can be used a filter when approving loans

### Data: LendingClub's Approved Loans after 2007. A fraction of the loan attributes are present. 

Models were developed from the variables that were available BEFORE the origination of the loan. Caution was taken to avoid certain variables that may have directly or indirectly been effected by the loan repayment performance.

Upon some resarching, the <b>'Does not meet the credit policy.'</b> phrase present in some loan statues does not seem to add any clear information. The credit policies change constantly. However, the succesfully repaid loan rate is significantly lower for such loans. 

In this case, we do have an abundance of data (entries vs attributes), and since loans with that phrase constitute a very small portion of the data, it is best to remove those loans from the analysis.  

```python
df = df[df.loan_status.apply(lambda x: not str(x).startswith('Does') )]
```

Alternatively, we could also treat them by ignoring that phrase: 

```python
if False: 
    #Drop the 'does not meet the credit policy' from 'status'. 
    df['loan_status'] = df.loan_status.apply(lambda x: str(x).split(':')[1] if str(x).startswith('Does') else str(x))
```

```python
df.loan_status.unique()
```


Since the objective is come up with a model that predicts loan outcome, data to be used in the analysis should contain the loans that we know the outcome for. Loans with statuses <b>['Current', 'In Grace Period', 'Late (16-30 days)']</b> can eventually go either way. 

Filter loans that are in <b>'Fully Paid', 'Late (31-120 days)', 'Charged Off'</b> and <b>'Default'</b> statuses: 

```python
df = df[df.loan_status.isin(['Fully Paid', 'Late (31-120 days)', 'Charged Off', 'Default'])]
```

Even though it makes intuitive sense to include loans that are <b>"31-120 days late"</b> in the Charged Off category, we cannot infer the chances of those loans' finalizing as Fully Paid since we do not have repayment behavior info for any loans except the total amount of payments made thus far. 

We could investigate them later, but let us excude these <b>"late loans"</b> for now. 

```python
# Filter out late loans for now. 
df_late = df.query('loan_status == "Late (31-120 days)"')
df      = df.query('loan_status != "Late (31-120 days)"' )
```


Loan Outcome: <b>"Success"</b> if Fully Paid, <b>"Failed"</b> if Charged Off or Defaulted

```python
df['Loan_Outcome'] = df.loan_status.apply(lambda x: 'Success' if x=='Fully Paid' else 'Failed')
```

### Split data into training, test and validation sets prior to any analysis (e.g. exploratory, modeling)

```python
# Make a Train-Test-Valid: 60-20-20 split
xk = np.arange(3)
pk = (0.6, 0.2, 0.2)
custm = sp.stats.rv_discrete(name='custm', values=(xk, pk))
class_label = custm.rvs(size=len(df))

df_train = df.iloc[class_label==0]
df_test  = df.iloc[class_label==1]
df_valid = df.iloc[class_label==2]
```

The revolving utilization rate can be considered a risk factor: 

```python
# 'revol_util'

# Remove the percent sign first
df_train['revol_util'] = df_train.revol_util.apply(lambda x: float(str(x).split('%')[0]) )
df_test['revol_util']  = df_test.revol_util.apply(lambda x: float(str(x).split('%')[0]) )
df_valid['revol_util'] = df_valid.revol_util.apply(lambda x: float(str(x).split('%')[0]) )

#Impute NANs with median
df_train['revol_util'][df_train.revol_util.isnull()] = df_train.revol_util.median(skipna=True)
df_test['revol_util'][df_test.revol_util.isnull()]   = df_train.revol_util.median(skipna=True)
df_valid['revol_util'][df_valid.revol_util.isnull()] = df_train.revol_util.median(skipna=True)

# Standardize
standardizer_revUtil = preprocessing.StandardScaler()
standardizer_revUtil.fit(np.array(df_train['revol_util']))
revUtil_standardized_train = standardizer_revUtil.transform(np.array(df_train['revol_util']))
revUtil_standardized_test  = standardizer_revUtil.transform(np.array(df_test['revol_util']))
revUtil_standardized_valid = standardizer_revUtil.transform(np.array(df_valid['revol_util']))

df_revUtil_std_train = pd.DataFrame(data=revUtil_standardized_train, columns=['revol_util_std'], index = df_train.index)
df_revUtil_std_test  = pd.DataFrame(data=revUtil_standardized_test,  columns=['revol_util_std'], index = df_test.index)
df_revUtil_std_valid = pd.DataFrame(data=revUtil_standardized_valid, columns=['revol_util_std'], index = df_valid.index)

df_train = df_train.join(df_revUtil_std_train)
df_test  = df_test.join(df_revUtil_std_test)
df_valid = df_valid.join(df_revUtil_std_valid)
```

### Get estimates of performance metrics on the boostrap samples of the validation set:

```python
from sklearn.utils import resample
```

```python
n_samples = 100 

gross_profit     = np.zeros((n_samples,))
net_gross_profit = np.zeros((n_samples,))
filter_roi       = np.zeros((n_samples,))
filter_roi_net   = np.zeros((n_samples,))
```

```python
for i in range(n_samples):
    samp_index = resample(y_valid.index)
    
    y_samp = y_valid.ix[samp_index]
    X_samp = X_valid.ix[samp_index]
    
    #ipdb.set_trace()
    
    y_samp_pred = models['gbm']['classifier'].predict(X_samp)
    
    gross_profit[i]     = np.sum(df_valid['profit'].ix[y_samp[y_samp_pred==0].index])
    net_gross_profit[i] = gross_profit[i] - np.sum(df_valid['profit'].ix[y_samp.index])
    
    filter_roi[i]     = np.sum(df_valid['total_pymnt'].ix[y_samp[y_samp_pred==0].index]) / float(np.sum(df_valid['funded_amnt'].ix[y_samp[y_samp_pred==0].index]))  - 1
    filter_roi_net[i] = filter_roi[i] -  (np.sum(df_valid['total_pymnt'].ix[y_samp.index]) / float(np.sum(df_valid['funded_amnt'].ix[y_samp.index])) - 1)
    
    filter_roi[i]     *= 100.0
    filter_roi_net[i] *= 100.0
    
    #ipdb.set_trace()
```
